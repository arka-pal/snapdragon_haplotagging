## Creating _posfiles_ with 100kb stitching regions around each chosen KASP marker

````bash
# mkdir -p ~/snap_hap/impute/stitch_tests/Chr{1..8}/stitch_regions/posfile_v2_filteredVCFs
````

_posfile_ is created using the script: `~/snap_hap/_scripts/bash/extract_posfile.sh`

````bash
## Initiate variables
WORKDIR=/nfs/scistore18/bartogrp/apal/snap_hap/impute/stitch_tests
vcfDIR=/nfs/scistore18/bartogrp/apal/snap_hap/variants/vcf_bcftools_Am_all
coords=$WORKDIR/coords/coords_for_stitch_manualFilter_100kb.bed

## Create posfile
while read chr chromStart chromEnd LocusName snpID;
do
    inVCF=$vcfDIR/$chr/Am_all_bcftools_${chr}_SnpGap5_biSNPs_filtered-DP_500-7732_QUAL20_MQ30.vcf.gz
    outfile=$WORKDIR/${chr}/stitch_regions/posfile_v2_filteredVCFs/${snpID}_${chromStart}_${chromEnd}-${LocusName}.pos
    echo -e "\n$snpID $chr $chromStart $chromEnd $LocusName"
    echo inVCF: $inVCF
    echo posfile: $outfile
    time bash ~/snap_hap/_scripts/bash/extract_posfile.sh $chr $chromStart $chromEnd $inVCF $outfile
done < $coords
````

````bash
## No. of SNPs in each posfile:
echo -e "No. of SNPs in 100kb region around chosen KASP markers\n"
for posfile in $WORKDIR/Chr?/stitch_regions/posfile_v2_filteredVCFs/*
do
    echo -e "`cat $posfile | wc -l` \t ${posfile##*/}"
done
````

## Run STITCH on SLURM

__NB:__ If running with `--plotAfterImputation=TRUE --plotHapSumDuringIterations=TRUE --plot_shuffle_haplotype_attempts=TRUE` only 1 job can run at a time in a particular partition since `xvfb-run: error: Xvfb failed to start`.  

For no plots, start with `srun STITCH.R <options>`  
For plots, start with `xvfb-run STITCH.R <options>`
